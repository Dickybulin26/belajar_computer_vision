### Matrix Multiplication (dot product)

dot_product.PNG

- hanya dapat dilakukan jika dan hanya jika jumlah kolom pada matrix kiri sama dengan jumlah baris pada matrix kanan. <br>

***(M. n)(n.m)***
untuk,  ***(M. n)***  merupakan ukuran matrix dibagian kanan, ***(n.m)***  merupakan ukuran matrix dibagian kiri.
dan akan menghasilkan matrix dengan ukuran  ***(M.m)***

- contoh :

$ X = \begin{bmatrix}
1 & 2 &1 \\
 3&0  &1
\end{bmatrix} $

- size X adalah 2,3

$ Y = \begin{bmatrix}
0 &3 \\
 5&1 \\
 7&3
\end{bmatrix} $

- size Y adalah 3,2

- sehingga kedua matrix diatas dapat dilakukan operasi dot untuk size (2,3)(3,2)

````python
X = np.array([[1,2,1],[3,0,1]])
Y  = np.array([[0,3,4,1],[5,1,4,1],[7,3,4,1]])
````

````python
X.shape, Y.shape
````



```
2,3 -- 3,4
```



````python
np.dot(X, Y)
````

````python
# (3,4)(2,3)
# np.dot(Y, X) # akan error
````

````python
# (3,4)(2,3)
np.dot(Y.T, X.T)
````
